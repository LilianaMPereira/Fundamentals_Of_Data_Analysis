### 3. Exploring Numpy Random Functionality
In this part of the work I will delve into the capabilities of Numpy's random module, which allows us to generate various types of random data.

#### 3.1 Understanding Numpy's Random

NumPy, also known as Numerical Python, is a crucial package for scientific and numerical computations in Python. It offers multidimensional arrays, specialized objects, and fast operations for manipulating arrays. Many Python tools for data analysis and machine learning, like pandas and scikit-learn, are built on top of NumPy. Its random module, numpy.random, is akin to Python's standard random library, but designed for NumPy arrays. It generates arrays of random numbers from various probability distributions, support tasks like data simulation and testing methods.

NumPy arrays, called ndarrays, can be constructed using functions like array, zeros, ones, and more. They possess attributes like shape, dtype, ndim, and size. The random module within NumPy can generate pseudorandom numbers, important for simulations. These numbers are based on deterministic algorithms, mimicking randomness.

Randomness has numerous applications in gaming, gambling, science, statistics, and machine learning. In data analytics and machine learning, random sampling from datasets is vital for method testing, algorithm evaluation, and cross-validation. The scikit-learn machine learning package relies on numpy.random for tasks like data splitting and cross-validation. This randomness helps evaluate algorithm performance on different subsets of data.

The random module in Numpy is a robust feature that provides an efficient means of producing pseudo-random numbers. While these numbers are not completely random, they are generated in a way that mimics randomness for practical use. The module employs a set of algorithms known as pseudorandom number generators (PRNGs) to create sequence of numbers that exhibit random characteristics.

The PRNG process operates starting with an initial value known as the seed. Through mathematical operations, a sequence of numbers is produced from this seed. It is important to emphasize that if the same seed is used, the same sequence of numbers will always be generated. This deterministic feature is extremely valuable as it allows for consistent results across experiments and simulations.

#### 3.2 Generating Random Numbers
Numpy's random module offers a range of functions to generate different types of random numbers. These include:

- `numpy.random.rand()`: Generates random numbers from a uniform distribution between 0 and 1.
- `numpy.random.randint()`: Generates random integers within a specified range.
- `numpy.random.random()`: Similar to `numpy.random.rand()`, but generates floating points between 0 and 1.
- `numpy.random.randn()`: Generates random numbers from a standard normal distribution (mean 0, standard deviation 1).

```python
# Generating Random Numbers with Numpy
random_uniform = np.random.rand(5) 
random_integers = np.random.randint(1, 7, size=5) 
random_floats = np.random.random(5)  
random_normal = np.random.randn(5)  

print("Random Uniform:", random_uniform)
print("Random Integers:", random_integers)
print("Random Floats:", random_floats)
print("Random Normal:", random_normal)
```

These sets of numbers represent various types of randomly generated distributions: the first one has values evenly spaced between 0 and 1, the second consists of integers from 1 to 6 simulating dice rolls, the third includes decimal numbers between 0 and 1 with varying distribution, and the last one follows a normal distribution centered around 0, mimicking a natural pattern of clustering.

In order to be able to check it in a graph, was increased the sample quantity.

```python
# Generating Random Numbers with Numpy
random_uniform = np.random.rand(1000) 
random_integers = np.random.randint(1, 7, size=1000)
random_floats = np.random.random(1000)  
random_normal = np.random.randn(1000) 

# Plotting the random number distributions
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(random_uniform, bins=20, alpha=0.7)
plt.title('Random Uniform Distribution')

plt.subplot(2, 2, 2)
plt.hist(random_integers, bins=6, alpha=0.7, rwidth=0.8)
plt.title('Random Integers Distribution')

plt.subplot(2, 2, 3)
plt.hist(random_floats, bins=20, alpha=0.7)
plt.title('Random Floats Distribution')

plt.subplot(2, 2, 4)
plt.hist(random_normal, bins=20, alpha=0.7)
plt.title('Random Normal Distribution')

plt.tight_layout()
plt.show()
```

The larger the sample size, the smoother the normal distribution.

```python
# Another Example

# Generate a 1D array of 5 random integers between 0 and 9
integer_array = np.random.randint(low=0, high=10, size=5)
print("1D Array of Random Integers:\n", integer_array)

# Generate a 1D array of 5 random floating-point numbers between 0 and 1
float_array = np.random.random(size=5)
print("\n1D Array of Random Floats:\n", float_array)

# Generate a 2D array of shape (3, 4) with random integers between 0 and 9
result = np.random.randint(low=0, high=10, size=(3, 4))
print("\n2D Array of Random Integers:\n", result)

# Plotting the random number distributions
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.hist(integer_array, bins=10, alpha=0.7)
plt.title('Random Integers Distribution')

plt.subplot(2, 2, 2)
plt.hist(float_array, bins=10, alpha=0.7)
plt.title('Random Floats Distribution')

plt.subplot(2, 2, 3)
plt.hist(result.flatten(), bins=10, alpha=0.7)
plt.title('Random Integers in 2D Array Distribution')

plt.tight_layout()
plt.show()
```

#### 3.3 Seed for Reproducibility

A "seed" is an initial value used to start a random number generator. When a generator is seeded, it produces a sequence of apparently random numbers that remain consistent as long as the same seed is used. This reproducibility is essential for maintaining consistent results in experiments, simulations, and analyses. By setting a seed, you can control the randomness and ensure that others can replicate your work with the same random outcomes.

```python
# Seed for Reproducibility
seed_value = 45
np.random.seed(seed_value)

print("\nGenerating random numbers with a seed of", seed_value, ":")

# Generating random integers between 1 and 100 
random_int_with_seed = np.random.randint(1, 101)
print("Random integer with seed:", random_int_with_seed)

# Generating random numbers from a standard normal distribution 
random_normal_with_seed = np.random.randn()
print("Random number from standard normal distribution with seed:", random_normal_with_seed)

# Generating an array of random floats 
random_array_with_seed = np.random.rand(5)
print("Array of 5 random floats with seed:", random_array_with_seed)
```

```python
# Plotting the random number distributions
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.hist([random_int_with_seed], bins=20, alpha=0.7)
plt.title('Random Integer Distribution')

plt.subplot(2, 2, 2)
plt.hist([random_normal_with_seed], bins=20, alpha=0.7)
plt.title('Random Normal Distribution')

plt.subplot(2, 2, 3)
plt.hist(random_array_with_seed, bins=20, alpha=0.7)
plt.title('Random Floats Distribution')

plt.tight_layout()
plt.show()
```

Using the same seed value, the code ensures that the generated random numbers are consistent and reproducible, making it easier to compare and analyze results across different runs of the code.

#### 4.2 Simulating Rolling a Six-Sided Dice

Consider the simulation of a standard six-sided die roll, an exemplary illustration of randomness in action. Using Numpy's random integer generation, I recreate the act of rolling a die and observing the corresponding result. Each simulated roll represents one of the die's six faces. This hands-on process offers a tangible understanding of random event simulation, providing a foundation for comprehending more intricate simulations.


```python
# Simulating Rolling a Six-Sided Dice
def simulate_dice_roll():
    return np.random.randint(low=1, high=7)

# Simulate multiple dice rolls
num_simulations = 1000
simulated_rolls = [simulate_dice_roll() for _ in range(num_simulations)]

# Analyzing Dice Roll Statistics
roll_counts = [simulated_rolls.count(i) for i in range(1, 7)]
total_rolls = sum(roll_counts)
roll_probabilities = [count / total_rolls for count in roll_counts]

# Displaying results
for i in range(1, 7):
    print(f"Number {i} appeared {roll_counts[i-1]} times with probability {roll_probabilities[i-1]:.3f}")
```

#### 4.3 Analyzing Dice Roll Statistics

Beyond simulating rolls, the analysis of resulting data is crucial. This involves assessing face frequencies and calculating key metrics like mean, variance, and standard deviation. This quantification unveils the inherent randomness of dice throws. Such statistical insights validate simulation fairness and deepen our grasp of outcome probabilities.

Throughout this section, we guide you through dice roll simulation using Numpy, offering code examples for standard six-sided die rolls. We also unveil statistical analysis techniques for the simulated data, yielding insights into underlying randomness. This exploration facilitates practical insight into probability and randomness, showcasing their applications across a spectrum of simulated scenarios.

```python
# Colour scheme and styles for plots.
sns.set_theme(palette="Dark2", context="paper", font_scale=1.5)

# Better sized plots.
plt.rcParams['figure.figsize'] = (12, 8)


def simulate_dice_rolls(num_rolls, num_sides):
    """
    Simulate rolling a standard die with a given number of sides
    for a specified number of times.
    """
    rolls = [np.random.randint(1, num_sides + 1) for _ in range(num_rolls)]   
    roll_counts = {side: rolls.count(side) for side in range(1, num_sides + 1)}
    return roll_counts

# Test the function
rolls_data = simulate_dice_rolls(1000, 6)
print(rolls_data)

# Plot the roll distribution
plt.bar(rolls_data.keys(), rolls_data.values())
plt.title("Dice Roll Simulation")
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.xticks(list(rolls_data.keys()))
plt.show()
```

### 5. Visualizing Random Data

#### 5.1 Introduction to Data Visualization

In this section, I'll look at the dynamic domain of visualizing random data.

#### 5.2 Histograms of Random Numbers

Histograms are a fundamental visualization technique that illustrate the distribution of data. In the context of random numbers, histograms provide a visual summary of how often different values occur. By graphing the frequency of values into buckets, we can quickly understand the shape and distribution of the data distribution. Applying histograms to random numbers improves our understanding of their underlying patterns.

Throughout this section, we demonstrate how to create impactful visual representations of random data. I started by introducing the concept and importance of data visualization. Later, we explored how to build histograms to visualize the distribution of random numbers. Finally, scatterplots were used to visually analyze the results of the input simulations.

```python
# Simulating random data
num_histograms = 5  
data_points = 10000  
bins = 15  

# Generating random numbers for each histogram
random_data = [np.random.randn(data_points) for _ in range(num_histograms)]

# Creating subplots for multiple histograms
fig, axes = plt.subplots(1, num_histograms, figsize=(15, 8), sharey=True)

# Creating histograms for each subplot
for i in range(num_histograms):
    axes[i].hist(random_data[i], bins=bins, color='blue', alpha=0.7)
    axes[i].set_title(f'Histogram {i + 1}')
    axes[i].set_xlabel('Value')
    if i == 0:
        axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### 5.3 Scatter Plots for Dice Roll Simulations

Scatter plots are excellent for representing relationships between two variables. When applied to rolling data simulations, scatterplots allow us to explore potential patterns or trends that can emerge from simulated data. By plotting the results of multiple data releases, we can uncover visual insights into simulated process behavior.

```python
# Simulating dice roll data for multiple simulations
num_simulations = 4  
dice_rolls = [np.random.randint(low=1, high=7, size=100) for _ in range(num_simulations)]

# Creating subplots for multiple scatter plots
fig, axes = plt.subplots(num_simulations, 1, figsize=(8, 6*num_simulations), sharex=True)

# Creating scatter plots for each subplot
for i in range(num_simulations):
    axes[i].scatter(range(1, 101), dice_rolls[i], color='green', alpha=0.7)
    axes[i].set_title(f'Scatter Plot - Simulation {i + 1}')
    if i == num_simulations - 1:
        axes[i].set_xlabel('Roll Number')
    axes[i].set_ylabel('Dice Value')
    axes[i].set_xticks(range(1, 101, 10))

plt.tight_layout()
plt.show()
```

The code generated four simulations of dice rolls, with each simulation containing 100 rolls. Four vertically stacked scatterplots were created. Each scatterplot showed how the data values varied across the releases, with the x-axis indicating the number of the release (from 1 to 100) and the y-axis showing the obtained data value. Each scatter plot represented a different simulation, allowing the comparison of launch patterns between simulations.
