## 1. Normal Distribution Notebook

### Properties of the Normal Distribution

The normal distribution, also known as the Gaussian distribution, is characterized by several crucial properties that establish it as the fundamental model for numerous real-world phenomena.

- **Central Tendency:**
In a normal distribution, the mean, median, and mode are located at the center of the distribution. This alignment highlights the distribution's balanced nature and is one of the reasons why the normal distribution is often used as a measure of central tendency.

- **Bell Curve:**
The bell-shaped curve of the normal distribution is its distinctive characteristic. It is also known as the "normal curve" because of its common occurrence in many natural phenomena. This curve demonstrates that values closer to the mean occur more frequently, while values farther from the mean decrease in frequency.

- **Standard Deviation and the 68-95-99.7 Rule:**
The standard deviation of the normal distribution plays a crucial role in understanding the data distribution. The 68-95-99.7 rule provides a quick estimate of how data is distributed around the mean and facilitates better interpretation of data points.

- **Symmetry:**
The most significant feature of a normal distribution is its symmetry. The distribution curve is perfectly symmetrical around its mean, resulting in mirrored images on both sides. This symmetry indicates that the probability of observing equally spaced mean values is the same.

- **Ubiquitous in Nature:**
Normal distributions spontaneously appear in many real-world situations, including measurements of physical attributes, measurement errors, IQ scores, and more. This ubiquity is one of the reasons why the normal distribution is such a valuable tool in statistics.

- **Density Everywhere:**
Unlike other distributions with gaps or isolated points, the normal distribution exhibits continuous density. This means that every possible value in the distribution has a nonzero probability of occurrence.

- **Addition of Independent Variables:**
When two independent random variables are added, their sum follows a normal distribution. This property has significant implications for the synthesis of random events.

- **Parameter Definition Format:**
The shape of the normal distribution is entirely determined by the mean and standard deviation. Different combinations of these parameters lead to different normal distributions, each with its own characteristics.


## 2. Generating Numbers from a Normal Distribution using NumPy

### 2.1 Importing Libraries
Before I start generating random numbers, I import the necessary libraries. NumPy to generate numbers and Matplotlib to visualize distributions.


```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 2.1.1 The `np.random.normal()` Function

NumPy provides a function called `np.random.normal()` that allows us to generate random numbers following a normal distribution. This function takes three main arguments:

- **`loc` (Mean):** This parameter specifies the mean of the distribution. It determines the center around which the values are generated.

- **`scale` (Standard Deviation):** This parameter sets the standard deviation of the distribution. It determines the spread or dispersion of the values around the mean.

- **`size` (Number of Samples):** This parameter determines the number of random samples to generate.

Here we have an example scenario where we want to generate a set of random heights of individuals from a normal distribution:

```python
mean_height = 170  # Mean height in centimeters
std_dev_height = 8  # Standard deviation of height
num_samples = 1000  # Number of individuals in the sample

heights = np.random.normal(mean_height, std_dev_height, num_samples)
```

#### 3.3 Example: Simulating Heights

To simulate the heights of a population. We can use the `np.random.normal()` function to generate random heights based on known population statistics. By adjusting the mean and standard deviation, it-s possible simulate different populations or scenarios.

```python
mean_height = 170  # Mean height in centimeters
std_dev_height = 8  # Standard deviation of height
num_samples = 1000  # Number of individuals in the sample

# Generate random heights 
heights = np.random.normal(mean_height, std_dev_height, num_samples)

# Visualize the distribution 
plt.hist(heights, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Simulated Heights from a Normal Distribution')
plt.grid(True)
plt.show()
```

### 4. Visualizing the Normal Distribution using Matplotlib

#### 4.1 Visualizing the Generated Data

Visualizing the generated data is crucial to understanding its characteristics. Histograms and probability density function (PDF) plots are effective tools for this purpose.

- **Histogram:** A histogram divides the data into intervals (bins) and displays how many values fall into each bin. This provides insight into the shape and spread of the distribution.

- **PDF Plot:** The PDF plot illustrates the likelihood of different values occurring. It's a smooth curve that represents the density of the distribution.

#### 4.2 Histogram Plot

A histogram is a graphical representation of the distribution of data. It's widely used in statistics and data analysis to visually summarize the frequency or probability density of values within a dataset. Histograms provide insights into the underlying data distribution, allowing us to identify patterns, central tendencies, and variability.

##### 4.2.1 Key Concepts:

- **Bins:**
   Bins are intervals into which the data range is divided. The x-axis of the histogram represents these bins. Each bin represents a range of values, and the height of the bar above a bin represents the frequency or density of data points falling within that range.

- **Frequency vs. Density:**
   - **Frequency:** In a frequency histogram, the height of each bar represents the actual count of data points within a bin.
   - **Density:** In a density histogram, the area under the curve of each bin represents the relative proportion of data points within that bin. This is useful for comparing histograms with different numbers of data points or bin widths.

- **Shape and Interpretation:**
   The shape of a histogram can provide insights into the underlying data distribution:
   - **Symmetry:** A symmetric histogram often indicates a normal distribution or a symmetric distribution.
   - **Skewness:** An asymmetric histogram may indicate skewness, where the distribution is stretched in one direction.
   - **Modality:** The number of peaks in the histogram can indicate the number of modes (peaks) in the data distribution.

- **Central Tendency and Spread:**
   The central tendency of the data, such as the mean or median, can be observed by locating the center of the histogram. The spread or variability of the data can be inferred from the width of the distribution.

- **Data Outliers:**
   Outliers, or data points that deviate significantly from the rest of the data, can be identified by observing bars that are unusually tall or separate from the main body of the histogram.

##### 4.2.2 Use Cases of Histograms:

- **Data Exploration:**
  Histograms help to quickly understand the distribution of data, identify patterns, and outliers.

- **Comparing Distributions:**
  Comparing multiple histograms can reveal a differences and similarities between datasets.

- **Probability Analysis:**
  In probability theory, histograms could be used to estimate probabilities and cumulative distribution functions.

- **Model Assessment:**
  Histograms can be compared to theoretical distributions to assess how well a model fits the data.

Histograms are a fundamental tool in exploratory data analysis, as they provide a visual summary of data characteristics, helping analysts and researchers make informed decisions based on data distribution and patterns.

#### Examples of Histograms:

```python
# Parameters for the normal distribution
mean = 0
std_dev = 1
num_samples = 10000

# Generate random numbers 
random_numbers = np.random.normal(mean, std_dev, num_samples)

# Visualize the distribution 
plt.hist(random_numbers, bins=20, density=True, alpha=0.6, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Random Numbers from a Normal Distribution')
plt.grid(True)
plt.show()
```

```python
# Parameters for the normal distribution
mean = 0
std_dev_values = [0.5, 1, 1.5]  

# Generate random numbers 
plt.figure(figsize=(15, 8))
for std_dev in std_dev_values:
    samples = np.random.normal(mean, std_dev, 1000)
    plt.hist(samples, bins=30, density=True, alpha=0.6, label=f'Standard Deviation = {std_dev}')

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histograms of Normal Distributions with Varying Standard Deviations')
plt.legend()
plt.grid(True)
plt.show()
```

In this example, the influence of the standard deviation on the shape of the distribution is investigated. By changing the standard deviation values, we can see how the dispersion around the mean affects the resulting histograms. This comparison gives insight and a deeper understanding of how the data propagation impacts the distribution characteristics.

```python
# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate random numbers
samples = np.random.normal(mean, std_dev, 1000)

x = np.linspace(min(samples), max(samples), 100)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

plt.figure(figsize=(15, 8))
plt.hist(samples, bins=30, density=True, alpha=0.6, label='Histogram')
plt.plot(x, pdf, color='r', label='PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram and PDF of a Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

In this example, a probability density function (PDF) curve is superimposed on the histogram, drawing a clear contrast between empirical and theoretical distributions. The visual alignment between these two representations highlights of the degree to which the observed data conforms to the predicted normal distribution.

```python
# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate random numbers
samples = np.random.normal(mean, std_dev, 1000)

plt.figure(figsize=(15, 8))
plt.hist(samples, bins=10, density=True, alpha=0.6, color='b', label='Bins = 10')
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Bins = 30')
plt.hist(samples, bins=50, density=True, alpha=0.6, color='r', label='Bins = 50')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histograms with Different Bin Counts')
plt.legend()
plt.grid(True)
plt.show()
```

Here we have an example of varying the number of bins when constructing a histogram. Adjusting the number of compartments acts as a kind of lens, through which we can increase or reduce the detail of the representation and by examining histograms with different amounts of bins, we gain an understanding of how granularity affects how we perceive the distribution of data.

For data visualization work in different areas (Science, engineering, economics, etc.), Matplotlib's histogram graph stands out as a powerful tool for revealing the nuances of normal distribution. It is worth highlighting the flexibility and effectiveness of histogram graphs. They not only show the multifaceted nature of the normal distribution, but also facilitate an intuitive exploration of patterns, central tendencies and variability hidden in the data, we unlock valuable insights into the behavior and characteristics of the dataset.

#### 4.2 Probability Density Function (PDF) Plot

The Probability Density Function (PDF) is a fundamental concept in probability and statistics, it describes the probability of a continuous random variable assuming a specific values. In the context of normal distribution, the PDF provides a mathematical representation of how the data is distributed around the mean. The PDF graph is a smooth curve that complements the histogram graph. The histogram displays the distribution of an observed data, and in turn, the PDF curve provides an idealized representation of the distribution. It is especially useful for comparing data to a theoretical distribution, such as the normal distribution

#### 4.2.1 Key Points:

- **Mathematical Form:** For a normal distribution with mean μ and standard deviation σ, the PDF is given by:

$$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{1}{2} \left(\frac{x - \mu}{\sigma}\right)^2} $$


Where x is the variable value, μ is the mean, σ is the standard deviation, and e is the base of the natural logarithm.

- **Area Under the Curve:** The area under the PDF curve over a specific range corresponds to the probability that the random variable falls within that range. This makes it a powerful tool for probability analysis.

- **Symmetry:** The PDF curve of a normal distribution is symmetric and bell-shaped. It reaches its maximum at the mean and decreases as values move away from the mean.

- **Standardization:** By standardizing the distribution (Z-score transformation), you can transform any normal distribution into the standard normal distribution with mean 0 and standard deviation 1.

- **Comparing Empirical Data:** Overlaying the PDF curve onto a histogram allows you to compare observed data with the idealized distribution. A close alignment suggests that the data follows the normal distribution.


#### 4.2.2 Use Cases:

- **Model Fitting:** PDF plots help assess how well a theoretical distribution fits observed data.
- **Probability Estimation:** PDF plots allow for probability estimation within specific ranges.
- **Comparative Analysis:** PDF plots help compare data distributions with different parameters.

The Probability Density Function (PDF) plot is a powrful tool to visualize the theoretical distribution of data, especially in the context of the normal distribution, this provides insights into probabilities, standardirzation, and how data aligns with theoretical expectations. The PDF plot complements the histograms, offering more continuous perspective on data distribution.


### 4.3 Examples of Probability Density Function (PDF) Plot

```python
# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate x values
x = np.linspace(-5, 5, 1000)

# Calculate the PDF values
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Create the PDF plot
plt.figure(figsize=(15, 8))
plt.plot(x, pdf, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.grid(True)
plt.show()
```

In this example, we're visualizing the Probability Density Function (PDF) of a standard normal distribution. The PDF is a bell-shaped curve centered at the mean (μ), and the standard deviation (σ) controls the spread. As values of x deviate from the mean, the probability density changes. The PDF is symmetric around the mean, and the area under the curve integrates to 1, representing total probability.

```python
# Parameter for the exponential distribution
lambda_param = 0.5  # Rate parameter

# Generate x values
x = np.linspace(0, 10, 1000)

# Calculate the PDF values
pdf = lambda_param * np.exp(-lambda_param * x)

# Create the PDF plot
plt.figure(figsize=(15, 8))
plt.plot(x, pdf, color='r')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of an Exponential Distribution')
plt.grid(True)
plt.show()
```

This example illustrates the PDF of an exponential distribution. The exponential distribution models time between events in a Poisson process. The PDF decreases exponentially as \(x\) increases; the parameter \(λ\) controls the decay rate; Higher \(λ\) values lead to faster decay.

```python
# Parameters for the uniform distribution
a = 0  # Lower bound
b = 10  # Upper bound

# Generate x values
x = np.linspace(a - 2, b + 2, 10000)

# Calculate the PDF values
pdf = np.where((x >= a) & (x <= b), 2 / (b - a), 0)

# Create the PDF plot
plt.figure(figsize=(15, 8))
plt.plot(x, pdf, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of a Uniform Distribution')
plt.grid(True)
plt.show()
```

In this example, we can see the PDF of a uniform distribution. The uniform distribution has constant probability density between the lower bound \(a\) and the upper bound \(b\), and is 0 outside that range. All values inside \([a, b]\) have equal probability.
